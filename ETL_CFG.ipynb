```python
# Agregamos la columna ID con la inicial del servicio Streaming y sumamos el contenido de la columna Show_id
Dataframe_amazon ['id'] = ('a') + Dataframe_amazon['show_id']
Dataframe_disney ['id'] = ('d') + Dataframe_disney['show_id']
Dataframe_hulu ['id'] = ('h') + Dataframe_hulu['show_id']
Dataframe_netflix ['id'] = ('n') + Dataframe_netflix['show_id']
```

```python
# se reemplazan los valores nulos del campo rating por el string 'G'
Dataframe_amazon.rating.fillna('G', inplace= True)
Dataframe_disney.rating.fillna('G', inplace= True)
Dataframe_hulu.rating.fillna('G', inplace= True)
Dataframe_netflix.rating.fillna('G', inplace= True)
```

```python
#De haber fechas, deberán tener el formato AAAA-mm-dd
Dataframe_amazon["date_added"] = pd.to_datetime(Dataframe_amazon["date_added"], format='%B %d, %Y')
Dataframe_disney["date_added"] = pd.to_datetime(Dataframe_disney["date_added"], format='%B %d, %Y')
Dataframe_hulu["date_added"] = pd.to_datetime(Dataframe_hulu["date_added"], format='%B %d, %Y')
# netflix me da el error "time data ' August 4, 2017' does not match format '%B %d, %Y' (match)", por lo que se normaliza el formato de las fechas quitando los espacios innecesarios.
#Dataframe_netflix['date_added'] = Dataframe_netflix['date_added'].str.strip()
#Dataframe_netflix["date_added"] = pd.to_datetime(Dataframe_netflix["date_added"], format='%B %d, %Y')

```

```python
# netflix me da el error "time data ' August 4, 2017' does not match format '%B %d, %Y' (match)", por lo que se normaliza el formato de las fechas quitando los espacios innecesarios.#
Dataframe_netflix['date_added'] = Dataframe_netflix['date_added'].str.strip()
Dataframe_netflix["date_added"] = pd.to_datetime(Dataframe_netflix["date_added"], format='%B %d, %Y')
Dataframe_netflix
```

```python
# Ahora se debe concatenar los data frame para poder unificar la base de datos en minuscula.
Dataframe_streamings = pd.concat([Dataframe_amazon,Dataframe_disney,Dataframe_hulu,Dataframe_netflix])
Dataframe_streamings
```

```python
# Se convierte el campo *duration* en dos campos: *`duration_int`* y *`duration_type`* segun lo solicitado.
Dataframe_streamings[['duration_int', 'duration_type']] = Dataframe_streamings ['duration'].str.split(' ', n=-1, expand=True, regex=None)
Dataframe_streamings2= Dataframe_streamings[Dataframe_streamings.columns[:10].tolist() + ['duration_int', 'duration_type'] + Dataframe_streamings.columns[10:-2].tolist()]
Dataframe_streamings2
```

```python
#Se unifican los campos de texto en minúsculas, sin excepciones.
Dataframe_streamings2['id'] = Dataframe_streamings2['id'].str.lower()
Dataframe_streamings2['type'] = Dataframe_streamings2['type'].str.lower()
Dataframe_streamings2['title'] = Dataframe_streamings2['title'].str.lower()
Dataframe_streamings2['director'] = Dataframe_streamings2['director'].str.lower()
Dataframe_streamings2['cast'] = Dataframe_streamings2['cast'].str.lower()
Dataframe_streamings2['country'] = Dataframe_streamings2['country'].str.lower()
Dataframe_streamings2['rating'] = Dataframe_streamings2['rating'].str.lower()
Dataframe_streamings2['listed_in'] = Dataframe_streamings2['listed_in'].str.lower()
Dataframe_streamings2['description'] = Dataframe_streamings2['description'].str.lower()
Dataframe_streamings2['duration'] = Dataframe_streamings2['duration'].str.lower()
# Reviso el data frame final 
Dataframe_streamings2
```

```python
# exporto el archivo a un csv nuevo:
Dataframe_streamings2.to_csv("Dataframe_streamings2.csv", index=False)
```
